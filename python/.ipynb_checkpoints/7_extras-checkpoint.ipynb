# Analyzing SOFC data (part 6)

In this notebook, I will cover a few basic statistical tests that weren't included in the machine learning analysis. These are just meant to provide some surface level understanding of the data. I will also revisit the categorical data and make some plots to show their relationships with important parameters.

```python
from scipy import stats
import pandas as pd
from sofc_func import *
%matplotlib inline
```

```python
# Import using pandas
df = pd.read_csv("data.csv")

# Clean out rows where there is no parent A or parent B

data = df[pd.notnull(df['A_par']) & pd.notnull(df['B_par']) & pd.notnull(df['d_star']) & pd.notnull(df['k_star']) 
             & pd.notnull(df['e affinity(B)']) & pd.notnull(df['d-electron count (B)'])]

pd.options.mode.chained_assignment = None

data = features(df)

temps = [400,500,600,700,800,900,1000]
for i in temps:
    data['dk_'+str(i)] = pd.Series(dk_star(data,i), index=df.index)

f = df[pd.notnull(data['EA_A']) & pd.notnull(data['EA_B']) & pd.notnull(data['r_A'])
        & pd.notnull(data['r_B']) & pd.notnull(data['d_count_B']) & pd.notnull(data['avg_EA_A'])
        & pd.notnull(data['avg_EA_B']) & pd.notnull(data['avg_r_A']) & pd.notnull(data['avg_r_B'])
        & pd.notnull(data['avg_d_count_B']) & pd.notnull(data['dk_'+str(temps[0])]) ]

#Sort by avg_EA_B
f = f.sort(['avg_EA_B'])
```

##ANOVA:
An analysis of variance (ANOVA) procedure measures the variation within groups against the variation between them. This is a way to determine if groups of data are significantly different than one another.

I will look at differences in both **D\*k\* and electron affinity of the B-site cation** to see if groups are statistically significant with respect to these parameters.

```python
family = data.groupby('family')
material = data.groupby('measurement_type')
sample = data.groupby('sample_type')
```

###Family:

```python
g = data.groupby('family')

p = g.get_group('Perovskite')
dp = g.get_group('Double Perovskite')
rp = g.get_group('Ruddlesden-Popper')

print 'D*k* : '
print 'T  | p-value between families'
for t in temps:
    _p = p[pd.notnull(p['dk_'+str(t)])]['dk_'+str(t)]
    _dp = dp[pd.notnull(dp['dk_'+str(t)])]['dk_'+str(t)]
    _rp = rp[pd.notnull(rp['dk_'+str(t)])]['dk_'+str(t)]
    _f,_pval = stats.f_oneway(_p,_dp,_rp)
    print t, _pval

print '\nElectron affinity (B) p-value'
ea_p = p[pd.notnull(p['e affinity(B)'])]['e affinity(B)']
ea_dp = dp[pd.notnull(dp['e affinity(B)'])]['e affinity(B)']
ea_rp = rp[pd.notnull(rp['e affinity(B)'])]['e affinity(B)']
_f, _pval = stats.f_oneway(ea_p, ea_dp, ea_rp)
print _pval
```

###Sample type:

```python
g = data.groupby('sample_type')

cer = g.get_group('Ceramic')
thin = g.get_group('Thin Film')
sing = g.get_group('Single Crystal')

print 'T  | p-value between sample types'
for t in temps:
    _cer = cer[pd.notnull(cer['dk_'+str(t)])]['dk_'+str(t)]
    _thin = thin[pd.notnull(thin['dk_'+str(t)])]['dk_'+str(t)]
    _sing = sing[pd.notnull(sing['dk_'+str(t)])]['dk_'+str(t)]
    _f,_pval = stats.f_oneway(_cer,_thin,_sing)
    print t, _pval
    
print '\nElectron affinity (B) p-value'
ea_cer = cer[pd.notnull(cer['e affinity(B)'])]['e affinity(B)']
ea_thin = thin[pd.notnull(thin['e affinity(B)'])]['e affinity(B)']
ea_sing = sing[pd.notnull(sing['e affinity(B)'])]['e affinity(B)']
_f, _pval = stats.f_oneway(ea_cer, ea_thin, ea_sing)
print _pval
```

```python
g = data.groupby('measurement_type')

iedp = g.get_group('IEDP/SIMS')
ecr = g.get_group('ECR')
ebsp = g.get_group('EBSP')

print 'T  | p-value between measurement types'
for t in temps:
    _iedp = iedp[pd.notnull(iedp['dk_'+str(t)])]['dk_'+str(t)]
    _ecr = ecr[pd.notnull(ecr['dk_'+str(t)])]['dk_'+str(t)]
    _ebsp = ebsp[pd.notnull(ebsp['dk_'+str(t)])]['dk_'+str(t)]
    _f,_pval = stats.f_oneway(_iedp,_ecr,_ebsp)
    print t, _pval
    
print '\nElectron affinity (B) p-value'
ea_iedp = iedp[pd.notnull(iedp['e affinity(B)'])]['e affinity(B)']
ea_ecr = ecr[pd.notnull(ecr['e affinity(B)'])]['e affinity(B)']
ea_ebsp = ebsp[pd.notnull(ebsp['e affinity(B)'])]['e affinity(B)']
_f, _pval = stats.f_oneway(ea_iedp, ea_ecr, ea_ebsp)
print _pval
```

###ANOVA Conclusion:
Difference in material family is statistically significant, but earlier we found that it added nothing to the model. What gives? Remember that material family is likely very strongly correlated with electron affinity of the B-site cation (or at least it is in the data that was collected). It is certainly correlated with material properties. Therefore, this result makes sense, but it is the actual properties themselves rather than the structure of the materials (i.e. the family) that contributes to variation and affects diffusion.

Differences in sample type were statistically significant, especially at higher temperatures. However, this again may be due to the fact that thin films and single crystals were generally only used for layered perovskites. I will check out this correlation as well.

Measurement type, as expected, was not statistically significant for D\*k\*. However, it was significant for electron affinity.

##Material family vs (electron affinity, D\*k\*)

```python
g = data.groupby('family')

p = g.get_group('Perovskite')
dp = g.get_group('Double Perovskite')
rp = g.get_group('Ruddlesden-Popper')

#fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(16,6))
for t in [400, 700, 1000]:
    dk_p = p[pd.notnull(p['dk_'+str(t)])]['dk_'+str(t)]
    dk_dp = dp[pd.notnull(dp['dk_'+str(t)])]['dk_'+str(t)]
    dk_rp = rp[pd.notnull(rp['dk_'+str(t)])]['dk_'+str(t)]
    dk_p.name = 'Perovskite'
    dk_rp.name = 'Ruddlesden-Popper'
    dk_dp.name = 'Double Perovskite'
    
    x = pd.concat([dk_p, dk_rp, dk_dp], axis=1)
       
    fig = plt.figure(figsize=(12,5))  
    ax = plt.subplot(111)
    x.plot(kind='box', ax=ax)
        
    plt.title('ln(D*k*) at '+str(t)+'K')
    plt.ylabel('ln(D*k*)')
    plt.show()
    
    
    
ea_p = p[pd.notnull(p['e affinity(B)'])]['e affinity(B)']
ea_dp = dp[pd.notnull(dp['e affinity(B)'])]['e affinity(B)']
ea_rp = rp[pd.notnull(rp['e affinity(B)'])]['e affinity(B)']
ea_p.name = 'Perovskite'
ea_rp.name = 'Ruddlesden-Popper'
ea_dp.name = 'Double Perovskite'
   
fig = plt.figure(figsize=(12,5))  
ax2 = plt.subplot(1,1,1)
x = pd.concat([ea_p, ea_rp, ea_dp], axis=1)
x.plot(kind='box', ax=ax2)
plt.title('Electron affinity (B)')
plt.ylabel('EN-B')
  
plt.show()
```

Family does follow a trend against D\*k\* at all temperatures, with **D\*k\*: Perovskite > Ruddlesden-Popper > Double Perovskite**, with this effect diminishing at higher temperatures as all familes come closer together.

Electron affinity, on the other hand, follows the order of **EN-B: Ruddlesden-Popper > Double Perovskite > Perovskite**, which is not the same trend as for D\*k\*.

However, for D\*k\* there is very little variability for RP and DP families, and for EN-B there is virtually no variability. This suggests the latter correlation is unconvincing, despite the statistical significance of EN-B vs Family.

##Sample type vs (electron affinity, D\*k\*)

```python
g = data.groupby('sample_type')

p = g.get_group('Ceramic')
dp = g.get_group('Thin Film')
rp = g.get_group('Single Crystal')

#fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(16,6))
for t in [400, 700, 1000]:
    dk_p = p[pd.notnull(p['dk_'+str(t)])]['dk_'+str(t)]
    dk_dp = dp[pd.notnull(dp['dk_'+str(t)])]['dk_'+str(t)]
    dk_rp = rp[pd.notnull(rp['dk_'+str(t)])]['dk_'+str(t)]
    dk_p.name = 'Ceramic'
    dk_rp.name = 'Thin Film'
    dk_dp.name = 'Single Crystal'
    
    x = pd.concat([dk_p, dk_rp, dk_dp], axis=1)
       
    fig = plt.figure(figsize=(12,5))  
    ax = plt.subplot(111)
    x.plot(kind='box', ax=ax)
        
    plt.title('ln(D*k*) at '+str(t)+'K')
    plt.ylabel('ln(D*k*)')
    plt.show()
    
    
    
ea_p = p[pd.notnull(p['e affinity(B)'])]['e affinity(B)']
ea_dp = dp[pd.notnull(dp['e affinity(B)'])]['e affinity(B)']
ea_rp = rp[pd.notnull(rp['e affinity(B)'])]['e affinity(B)']
ea_p.name = 'Ceramic'
ea_rp.name = 'Thin Film'
ea_dp.name = 'Single Crystal'
   
fig = plt.figure(figsize=(12,5))  
ax2 = plt.subplot(1,1,1)
x = pd.concat([ea_p, ea_rp, ea_dp], axis=1)
x.plot(kind='box', ax=ax2)
plt.title('Electron affinity (B)')
plt.ylabel('EN-B')
  
plt.show()
```

This is much the same story as before, only with less significant correlations. There is zero variation in thin film samples, but based on the plots, it seems there is a strong correlation between ceramic/perovskite, thin film/RP, and single crystal/(RP,DP).

#Conclusion:
I have looked at ANOVA results for the three categories discussed in a previous notebook. The tests for sample type and family both show significant differences between groups, but with the overwhelming majority of variation coming from perovskite ceramic samples.

**While family may, in a sense, predict trends in D\*k\*, it does not explain any variation in the data. Therefore, material family alone should not be used to predict outcomes.**

Family and sample type both show trends against electron affinity that are different than the trend observed for D\*k\*. This likely has something to do with the fact that EN_B is predictive of D\*k\*, but by a polynomial rather than a linear relationship. 

In summary, the data in this analysis is not diverse enough to make clear predictions using sample type, family, or measurement type. However, the previous use of EN_B appeared a strong enough predictor of diffusion throughout a material, with a polynomial relationship. There is much more to be explained, which is likely a result of different experimental setups, measurement error, and differences in stoichiometries. 

Nevertheless, EN_B appears the strongest predictor of diffusion in SOFC material. The B-site cation in general appears a much stronger predictor than the A-site cation. More data is needed (especially of layered perovskites) to make more certain conclusions.

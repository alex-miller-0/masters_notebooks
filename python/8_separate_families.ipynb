#SOFC analysis (part 8)

##Separating perovskite and Ruddlesden-Popper data
After the previous analysis, it became clear to me that electron affinity should not affect diffusion in Ruddlesden-Popper materials. This is because diffusion in those materials occurs via an interstitial mechanism rather than the vacancy migration mechanism seen in perovskites and double perovskites.

Thus, I am splitting the data into perovskites/double perovskites and RP.

```python
from sofc_func import *
%matplotlib inline
```

```python
# Import using pandas
df = pd.read_csv("data.csv")

# Clean out rows where there is no parent A or parent B

data = df[pd.notnull(df['A_par']) & pd.notnull(df['B_par']) & pd.notnull(df['d_star']) & pd.notnull(df['k_star']) 
             & pd.notnull(df['e affinity(B)']) & pd.notnull(df['d-electron count (B)'])]

pd.options.mode.chained_assignment = None

data = features(df)
```

```python
temps = [400,500,600,700,800,900,1000]
for i in temps:
    data['dk_'+str(i)] = pd.Series(dk_star(data,i), index=data.index)
    data['d_'+str(i)] = pd.Series(arrhenius(data['d_star'].values, data['ead'].values, data['T'].values, i))
    data['k_'+str(i)] = pd.Series(arrhenius(data['k_star'].values, data['eak'].values, data['T'].values, i))

data = data[pd.notnull(data['EA_A']) & pd.notnull(data['EA_B']) & pd.notnull(data['r_A'])
        & pd.notnull(data['r_B']) & pd.notnull(data['d_count_B']) & pd.notnull(data['avg_EA_A'])
        & pd.notnull(data['avg_EA_B']) & pd.notnull(data['avg_r_A']) & pd.notnull(data['avg_r_B'])
        & pd.notnull(data['avg_d_count_B']) & pd.notnull(data['dk_'+str(temps[0])]) ]

#Sort by avg_EA_B
data = data.sort(['avg_EA_B'])
```

```python
per = data[data.family.isin(["Perovskite", "Double Perovskite"])]
rp = data[data.family == 'Ruddlesden-Popper']
```

##Perovskites and Double Perovskites
For these materials, through which oxide ions diffuse via vacancy migration, I will perform the same analysis as before.

```python
fam = pd.get_dummies(per['family'], prefix='family')
sam = pd.get_dummies(per['sample_type'], prefix='s')

per = per.drop('family', axis=1)
per = per.drop('sample_type', axis=1)

per = pd.concat([per, fam], axis=1)
per = pd.concat([per, sam], axis=1)


#Sort by avg_EA_B
per = per.sort(['avg_EA_B'])
```

```python
_features = per[ ['avg_EA_A', 'avg_r_A','avg_r_B', 'tol_factor', 'r_critical', 'avg_EA_B', 'avg_d_count_B', 'pB_par' ]]
y = per['dk_1000']
ft = add_feature(None, np.array([y]).T, _features, 5, 50)
```

```python
ftt = add_feature(ft, np.array([y]).T, _features, 0.1, 100)
```

Once again, avg_EA_B alone is the best feature subset (lowest MSE).

```python
X = per[['avg_EA_B']]
```

```python
def get_predictions(iterations, len_data, order, X, Y):
    predictions = []
    for cycle in range(iterations):
        cycle_predictions = []
        #lko = cross_validation.KFold(len_data, n_folds=k, shuffle=True)
        lko = cross_validation.LeavePOut(len_data, 1)
        for train_index, test_index in lko:
            x_train, x_test = X[train_index], X[test_index]
            y_train, y_test = Y[train_index], Y[test_index]
            model = make_pipeline(PolynomialFeatures(order), Ridge()).fit(x_train, y_train)
            cycle_predictions.append(model.predict(X))
        predictions.append(np.mean(np.array(cycle_predictions), axis=0))
    p = np.mean(np.array(predictions), axis=0)
    return p
```

```python
scores = []
stds = []

models = {}
d_star = {}
k_star = {}

iterations = 5

start = time.clock()
for i in temps:
    y = per['dk_'+str(i)]
    o = optimize_order(iterations,len(y), 1, np.array(X), np.array(y), True)
    score, std = get_scores(iterations, 1, o, len(y), np.array(X), np.array(y), True)
    scores.append(score)
    stds.append(std)
    print 'Temperature: '+ str(i)+ ' Polynomial order: '+ str(o) + ' Score: ' + str(score) + ' +/- ' + str(std)
    models[str(i)] = get_predictions(iterations, len(y), o, np.array(X), np.array(y))
print 'Executed in ' + str(time.clock()-start) + 's'
```

###Plotting vs EA_B:

```python
labels = []
plt.figure(figsize=(12,8))
for temp in temps:
    plt.plot(X['avg_EA_B'], models[str(temp)], label=str(temp)+'K', linewidth=2)
    plt.ylabel('ln(D*k*)', fontsize=18)
    plt.xlabel('Electron affinity of B-cation (eV)', fontsize=18)
    #plt.title('ln(D*k*) as a function of temperature', fontsize=24)
    labels.append(str(temp)+'K')
plt.legend(labels, fontsize=16)
plt.show()
```

Wow! Removing Ruddlesden-Popper materials from the landscape turns this into a roughly linear relationship, with steeper slopes at lower temperatures. At high temperatures, the relationship begins curving and is best represented as a 3rd or 5th order polynomial function, but those are regularized and are still not far from linear.

###Adding dummy variables:
I will run the regressions for both EA_B and the dummy variables. This is for perovskites/double perovskites so family_Ruddlesden-Popper will be removed. No single crystal samples exist for these familys so that variable will also be removed.

```python
X = per[['avg_EA_B', 
         's_Ceramic','s_Thin Film', 
         'family_Perovskite','family_Double Perovskite']]
```

```python
models = {}
d_star = {}
k_star = {}

iterations = 5

start = time.clock()
for i in temps:
    y = per['dk_'+str(i)]
    o = optimize_order(iterations,len(y), 1, np.array(X), np.array(y), True)
    #score, std, r2, r2_std = get_scores(iterations, 1, o, len(y), np.array(X), np.array(y), True)
    models[str(i)] = get_predictions(iterations, len(y), o, np.array(X), np.array(y))
print 'Executed in ' + str(time.clock()-start) + 's'
```

```python
from sklearn.metrics import r2_score
r2s = []
for i in temps:
    r2s.append(r2_score(models[str(i)], per['dk_'+str(i)]))

plt.plot(temps, r2s, marker='^')
plt.xlim(0.9*np.min(temps), 1.1*np.max(temps) )
plt.xlabel('Temperature (K)', fontsize=18)
plt.ylabel(r"$R^2$", fontsize=18)
#plt.title('Model fit vs temperature')
plt.show()
```

The coefficients of determination (R^2) increase with higher temperature. The model, which includes the dummy variables, captures 85-90% of the variation in ln(D\*k\*) at 800-1000K. The model is not as good a predictor at lower temperatures, but still captures the majority of variation.

#Conclusion:
Perovskite data can be modeled with average electron affinity of the B-site cation as the only predictor. This model is roughly linear, but at high temperatures shows some polynomial character. Adding dummy variables corresponding to material family and sample type, up to 85% of the variability in the data can be explained by the model. For lower temperatures, this number is closer to 70%. The remainder of variation comes from extrapolation error and experimental inconsistencies.

Ruddlesden-Popper data cannot be modeled with sufficient confidence. More data is needed to make an assessment of features that lead to fast ion diffusion in Ruddlesden-Popper materials.

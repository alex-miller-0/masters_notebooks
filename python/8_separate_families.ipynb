##Separating perovskite and Ruddlesden-Popper data
After the previous analysis, it became clear to me that electron affinity should not affect diffusion in Ruddlesden-Popper materials. This is because diffusion in those materials occurs via an interstitial mechanism rather than the vacancy migration mechanism seen in perovskites and double perovskites.

Thus, I am splitting the data into perovskites/double perovskites and RP.

```python
from sofc_func import *
%matplotlib inline
```

```python
# Import using pandas
df = pd.read_csv("data.csv")

# Clean out rows where there is no parent A or parent B

data = df[pd.notnull(df['A_par']) & pd.notnull(df['B_par']) & pd.notnull(df['d_star']) & pd.notnull(df['k_star']) 
             & pd.notnull(df['e affinity(B)']) & pd.notnull(df['d-electron count (B)'])]

pd.options.mode.chained_assignment = None

data = features(df)
```

```python
temps = [400,500,600,700,800,900,1000]
for i in temps:
    data['dk_'+str(i)] = pd.Series(dk_star(data,i), index=data.index)
    data['d_'+str(i)] = pd.Series(arrhenius(data['d_star'].values, data['ead'].values, data['T'].values, i))
    data['k_'+str(i)] = pd.Series(arrhenius(data['k_star'].values, data['eak'].values, data['T'].values, i))

data = data[pd.notnull(data['EA_A']) & pd.notnull(data['EA_B']) & pd.notnull(data['r_A'])
        & pd.notnull(data['r_B']) & pd.notnull(data['d_count_B']) & pd.notnull(data['avg_EA_A'])
        & pd.notnull(data['avg_EA_B']) & pd.notnull(data['avg_r_A']) & pd.notnull(data['avg_r_B'])
        & pd.notnull(data['avg_d_count_B']) & pd.notnull(data['dk_'+str(temps[0])]) ]

#Sort by avg_EA_B
data = data.sort(['avg_EA_B'])
```

```python
per = data[data.family.isin(["Perovskite", "Double Perovskite"])]
rp = data[data.family == 'Ruddlesden-Popper']
```

```python
def get_predictions(iterations, len_data, k, order, X, Y):
    predictions = []
    for cycle in range(iterations):
        cycle_predictions = []
        lko = cross_validation.KFold(len_data, n_folds=k, shuffle=True)
        for train_index, test_index in lko:
            x_train, x_test = X[train_index], X[test_index]
            y_train, y_test = Y[train_index], Y[test_index]
            model = make_pipeline(PolynomialFeatures(order), Ridge()).fit(x_train, y_train)
            cycle_predictions.append(model.predict(X))
        predictions.append(np.mean(np.array(cycle_predictions), axis=0))
    p = np.mean(np.array(predictions), axis=0)
    return p
```

```python
scores = []
stds = []
X = per[['avg_EA_B']]
models = {}
d_star = {}
k_star = {}

iterations = 600

start = time.clock()
for i in temps:
    y = per['dk_'+str(i)]
    k = optimize_k(iterations, len(y), np.array(X), np.array(y), True)
    o = optimize_order(iterations,len(y), k, np.array(X), np.array(y), True)
    score, std = get_scores(iterations, k, o, len(y), np.array(X), np.array(y), True)
    scores.append(score)
    stds.append(std)
    print 'Temperature: '+ str(i)+ ' Polynomial order: '+ str(o) + ' Score: ' + str(score) + ' +/- ' + str(std)
    models[str(i)] = get_predictions(iterations, len(y), k, o, np.array(X), np.array(y))
print 'Executed in ' + str(time.clock()-start) + 's'
```

```python
labels = []
plt.figure(figsize=(12,16))
for temp in temps:
    plt.plot(X, models[str(temp)], label=str(temp)+'K', linewidth=3)
    plt.ylabel('ln(D*k*)', fontsize=18)
    plt.xlabel('Electron affinity of B-cation (eV)', fontsize=18)
    plt.title('ln(D*k*) as a function of temperature', fontsize=24)
    labels.append(str(temp)+'K')
plt.legend(labels, fontsize=16)
plt.show()
```

Wow! Removing Ruddlesden-Popper materials from the landscape turns this into a roughly linear relationship, with steeper slopes at lower temperatures. At high temperatures, the relationship begins curving and is best represented as a 2nd or 4th order polynomial function.
